```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
```

#Nomor 1

```python
array1 = np.array(img.imread('banana.jpg'))
plt.imshow(array1)
```

```python
array1
```

```python
zoom = np.zeros([int(len(array1)/2), int(len(array1[0])/2), 3])
for x in range(0, len(zoom)):
  for y in range(0, len(zoom[0])):
    for z in range(0, 3):
      result = array1[x*2][y*2][z]//4 + array1[x*2][(y*2)+1][z]//4 + array1[(x*2)+1][y*2][z]//4 + array1[(x*2)+1][(y*2)+1][z]//4
      zoom[x][y][z] = result
      result = 0
zoom = np.round(zoom).astype(int)
zoom
```

```python
plt.imshow(zoom)
```

#Nomor 3

```python
image = cv2.imread('kabut.png')

kernel1 = np.array([[0, -1,  0],
                   [-1,  5, -1],
                    [0, -1,  0]])

identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)

# We should get the same image
plt.imshow(identity)
```

#Nomor 5

```python
image = cv2.imread('phone_noise.jpg') # reads the imag
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
new_image = cv2.medianBlur(image, 9)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Medium')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
image = cv2.imread('phone_noise.jpg') # reads the imag
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
new_image = cv2.blur(image,(9, 9))
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Mean')
plt.xticks([]), plt.yticks([])
plt.show()
```

#<h1>Nomor 3</h1>

```python
im = cv2.imread("/content/coints.png")
im2 = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret2,th2 = cv2.threshold(im2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th2, cmap="gray")
plt.title("Gambar awal")
```

```python
image_r = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
im2 = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
plt.imshow(im2[:, :, 2], cmap="hsv")
plt.title("Gambar dalam HSV")
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))
im = ax[0].imshow(im2[:,:,2],cmap='hsv')
fig.colorbar(im,ax=ax[0])
ax[0].set_title('Komponen V',fontsize=15)

lower_mask = im2[:,:,2] >= 81
upper_mask = im2[:,:,2] < 260
mask = upper_mask*lower_mask

red = image_r[:,:,0]*mask
green = image_r[:,:,1]*mask
blue = image_r[:,:,2]*mask
mask2 = np.dstack((red,green,blue))
ax[1].imshow(mask)
ax[2].imshow(mask2)

ax[1].set_title('Mask',fontsize=15)
ax[2].set_title('Gambar Akhir',fontsize=15)
plt.tight_layout()
plt.show()
```

```python
im2 = cv2.cvtColor(mask2, cv2.COLOR_BGR2GRAY)
ret2,th2 = cv2.threshold(im2,0,255,cv2.THRESH_BINARY)
plt.figure(figsize=(10,10))
plt.xticks(np.arange(0, im2.shape[1], 25))
plt.yticks(np.arange(0, im2.shape[0], 25))
plt.imshow(th2, cmap="gray")
plt.title("Hasil Akhir")
```

#<h1>Nomor 4</h1>

```python
image = cv2.imread("/content/peppers.tiff")
image_r = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_r)
plt.title("Gambar Awal")
```

```python
twoDimage = image_r.reshape((-1,3))
twoDimage = np.float32(twoDimage)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 0.85)
k = 3
ret,label,center=cv2.kmeans(twoDimage,k,None,criteria,5,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((image_r.shape))
plt.imshow(result_image)
plt.title("Gambar pasca K-means")
```

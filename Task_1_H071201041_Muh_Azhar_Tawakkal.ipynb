#<h1>Nomor 8<h1>

<h2>Rotasi citra lena_color.bmp dengan sudut 30 dengan sudut putar titik tengah citra</h2>

```python
array8 = np.array(img.imread('lena_color.gif'))
plt.imshow(array8)
```

```python
array8
```

```python
rotate_center = [255,255]
x_center = rotate_center[0]
y_center = rotate_center[1]
rad = np.deg2rad(30)
rad
```

```python
array_result = np.zeros([len(array8), len(array8[0]), 4])
for x in range(0, len(array8)):
  for y in range(0, len(array8[0])):
    coor_x = ((x-y_center)*np.cos(rad) - (y-x_center)*np.sin(rad))+x_center
    coor_y = ((x-y_center)*np.sin(rad) + (y-x_center)*np.cos(rad))+y_center
    if coor_x == x and coor_y == y :
      print(x,y)
    if coor_x <= 512 and  coor_x >= 0 and coor_y <= 512 and  coor_y >= 0 :
      array_result[int(coor_x)][int(coor_y)] = array8[x][y]
hasil_rotate = np.round(array_result).astype(int)
hasil_rotate
```

```python
plt.imshow(hasil_rotate)
```

#<h1>Nomor 10<h1>

<h2>Perbesar citra girl-warna.bmp menjadi 512  512</h2>

```python
array2 = np.array(img.imread('girl-warna.bmp'))
plt.imshow(array2)
```

```python
array2
```

```python
zoom = np.zeros([len(array2)*2, len(array2[0])*2, 4])
a,b = 0,0
for x in range(0, len(array2)):
  for y in range(0, len(array2[0])):
    if a == len(zoom):
      continue
    if b == len(zoom[0]):
      continue
    zoom[a][b] = array2[x][y]
    zoom[a][b+1] = array2[x][y]
    zoom[a+1][b] = array2[x][y]
    zoom[a+1][b+1] = array2[x][y]
    b = b+2
  a = a+2
  b = 0
zoom = np.round(zoom).astype(int)
zoom
```

```python
plt.imshow(zoom)
```

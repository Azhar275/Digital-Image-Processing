```python
im = cv2.imread("/content/rice.png", 0)
array2 = np.array(im)
kernel_log = np.array(
[[ 0, 0, -1, 0,  0],
 [0,  -1,  -2,  -1, 0],
 [-1,  -2,  16,  -2, -1],
 [0,  -1,  -2,  -1, 0],
 [ 0, 0, -1, 0,  0]])
output3 = cv2.filter2D(im, -1, kernel_log)
plt.imshow(output3, cmap="gray")
```

Berdasarkan kedua kode, dapat dilihat perbedaan pada bagaimana kedua gambar menampilkan tepi


*   Gambar pertama mampu menampilkan tepi objek dengan baik, namun hasilnya tidak secerah gambar yang kedua. Gambar kedua memperlihatkan dengan sangat jelas tepi dari objeknya
*   Gambar kedua dapat mendeteksi titik titik pada gambar dibandingkan dengan gambar pertama.

Jadi, berdasarkan gambar tersebut, kode LoG dapat memperlihatkan tepi dan titik-titik dengan sangat baik dibandingkan kode Laplace. Laplace untuk gambar tersebut dapat digunakan jika hanya ingin sekedar melihat tepi dari objek, tanpa terlalu menonjolkan titik pada gambar


#Nomor 2

```python
im = cv2.imread("/content/boat.tiff", 0)
source = cv2.GaussianBlur(im, (3, 3), 0)

plt.figure(figsize=(10,10))

#Gambar asli
plt.subplot(2,2,1), plt.imshow(source, cmap='gray')
plt.title('Original')

#Sobel
kernel_sx = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
kernel_sy = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
img_sobelx = cv2.filter2D(source, -1, kernel_sx)
img_sobely = cv2.filter2D(source, -1, kernel_sy)
plt.subplot(2,2,2), plt.imshow(img_sobelx + img_sobely, cmap='gray')
plt.title('Sobel')

#Prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(source, -1, kernelx)
img_prewitty = cv2.filter2D(source, -1, kernely)
plt.subplot(2,2,3), plt.imshow(img_prewittx + img_prewitty, cmap='gray')
plt.title('Prewitt')

#Roberts
kernel_rx = np.array([[0,1],[-1,0]])
kernel_ry = np.array([[1,0],[0,-1]])
img_rx = cv2.filter2D(source, -1, kernel_rx)
img_ry = cv2.filter2D(source, -1, kernel_ry)
plt.subplot(2,2,4), plt.imshow(img_rx + img_ry, cmap='gray')
plt.title('Roberts')
```

Untuk citra boat, filter yang cocok digunakan menurut saya adalah filrer sobel. Hal ini disebabkan karena filter sobel dapat memperlihatkan lebih banyak tepi serta tepi yang ditampilkan juga lebih terang. <br><br>
Selain itu, karena citra boat memiliki banyak tepi yang harus dideteksi, maka opsi terbaik adalah filter sobel yang mampu menampilkan dengan sangat terang setiap edge
